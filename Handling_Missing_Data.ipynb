```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv('/content/pima-indians-diabetes.csv')
```

```python
print(data)
```

```python
#Seperating factors and target
factors = data.iloc[:, :-1]
target = data.iloc[:, -1]
```

```python
#Counting the number of missing values in each column of factors
missing = (factors == 0).sum()

print(missing)
```

```python
imp_mean = SimpleImputer(missing_values = 0 , strategy ='mean')
```

```python
#Applying fit
imp_mean.fit(factors)
```

```python
improved_data = imp_mean.transform(factors)
```

```python
improved_data = pd.DataFrame(improved_data, columns = factors.columns)
```

```python
data_imputed = pd.concat([improved_data, target], axis=1)
```

```python
data_imputed.to_csv('improveddata.csv' ,index =False)
```
